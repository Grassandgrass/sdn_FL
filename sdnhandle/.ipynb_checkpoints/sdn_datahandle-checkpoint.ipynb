```python
### 处理攻击数据
ddos_flow_dataset = pd.read_csv('./data/FlowStatsfile_ddos.csv',names = col_names)
```

```python
### 处理正常数据
flow_dataset = pd.read_csv('./data/FlowStatsfile.csv',names = col_names)
```

```python
flow_dataset = flow_dataset.drop(0)
flow_dataset = shuffle(flow_dataset) # 打乱数据 
flow_dataset = flow_dataset.iloc[0:10000]
flow_dataset.head(10)
```

```python
ddos_flow_dataset.iloc[:, 2] = ddos_flow_dataset.iloc[:, 2].str.replace('.', '')
ddos_flow_dataset.iloc[:, 3] = ddos_flow_dataset.iloc[:, 3].str.replace('.', '')
ddos_flow_dataset.iloc[:, 5] = ddos_flow_dataset.iloc[:, 5].str.replace('.', '')
flow_dataset.head(10)
```
